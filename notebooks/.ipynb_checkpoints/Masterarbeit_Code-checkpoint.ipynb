# Crawler immobilienscount24.de
Here the webpage immobilienscout24.de is crawled and the data is stored in a pandas data frame. For this, two larger functions are needed. 

### Get the apartment URLs
This function is used to download all the urls from the overview pages of immobilienscout24.de. If the archive of the website is the desired source, the argument 'archive' has to be given.

```python
def get_apartment_urls(source_type=''):
    columns = ['url']
    df = pd.DataFrame(columns=columns)
    url_list =[]
    time_beginning = datetime.datetime.now()
    if source_type == 'archive':
        pagelimit = 519
    else: 
        pagelimit = 190
    for i in range(1, pagelimit):
        if source_type == 'archive':
            url = 'https://www.immobilienscout24.de/Suche/S-T/P-{}/Wohnung-Miete/archiv/Umkreissuche/Berlin/-/229459/2511140/-/-/50?enteredFrom=result_list'.format(i)
        else:
            url = 'https://www.immobilienscout24.de/Suche/S-T/P-{}/Wohnung-Miete/Umkreissuche/Berlin/-/229459/2511140/-/-/50?enteredFrom=result_list'.format(i)
        r = requests.get(url)
        data = r.text
        soup = BeautifulSoup(data, "html5lib")
        urls = soup.find_all('a', href=re.compile(r'(expose)(.*?)[0-9]{2,}'))
        for x in urls:
            url_list.append(x['href'])
        sys.stdout.write('\r'+"parsing page # " + str(i) + "  " + "time passed: " + str(datetime.datetime.now() - time_beginning))

    url_list= pd.Series(url_list)
    df['url'] = url_list.values
    df['url'] = df['url'].str.split('#').str[0]
    df['url'] = 'https://www.immobilienscout24.de' + df['url']

    df = df.drop_duplicates()
    df = df.reset_index(drop=True)
    return df
```

### Function get_html_elements
Function to get elements from immobilienscout24.de. The arguments are the html code element as well as the desired class, and the third argument has to be set as soup each time. The elements are selected and then sent to the text_preprocessing function.

```python
def get_html_elements(element, html_class, soup):
    try:        
        html_element = soup.find(element, class_=html_class).get_text()
        html_element = text_preprocessing(html_element)
        return html_element
    except:
        return np.nan
    
```

### Function text_processing
The downloaded data is sent from the get_html_elements function to the text_processing function, were the text data is cleaned up a bit, but not completley. The complete preprocessing of the data frame follows later on.

```python
def text_preprocessing(element):
    try:
        element = str(element)
        element = element.replace('.', '')
        element = element.replace(',', '.')
        element = element.replace('€', '')
        element = element.replace('m²', '')
        element = element.replace('+', '')
        element = element.replace('Fotos ansehen', '')
        element = element.strip()
        return element
    except:
        return element
```

# Parser for CityGML data
In this section the CityGML .gml files are parsed and all interesting data is extracted. The extract_citygml() function needs the gml file as an input. To get additional data from the gml, an xpath can be set and looped through (see for height... and for poly...). The result is a pandas data frame containing the ID of the building, the ground surfaces (can be more than one per ID), the height of the building and the polygon values for the ground and roof surfaces.

```python
def extract_citygml(xml_data):
    with open(xml_data) as f:
        lines = f.readlines()
    lines
    lines[1] = '<CityModel xmlns:xal="xal" xmlns:gml="gml" xmlns:wtr="wtr" xmlns:app="app" xmlns:tex="tex" xmlns="citygml" xmlns:veg="veg" xmlns:dem="rel" xmlns:tran="tran" xmlns:bldg="bldg" xmlns:grp="grp" xmlns:tun="tun" xmlns:frn="frn" xmlns:brid="brid" xmlns:gen="gen" xmlns:xlink="xlink" xmlns:luse="luse" xmlns:xsi="xsi" xsi:schemaLocation="http://www.opengis.net/citygml/waterbody/2.0 http://schemas.opengis.net/citygml/waterbody/2.0/waterBody.xsd http://www.opengis.net/citygml/appearance/2.0 http://schemas.opengis.net/citygml/appearance/2.0/appearance.xsd http://www.opengis.net/citygml/texturedsurface/2.0 http://schemas.opengis.net/citygml/texturedsurface/2.0/texturedSurface.xsd http://www.opengis.net/citygml/vegetation/2.0 http://schemas.opengis.net/citygml/vegetation/2.0/vegetation.xsd http://www.opengis.net/citygml/relief/2.0 http://schemas.opengis.net/citygml/relief/2.0/relief.xsd http://www.opengis.net/citygml/transportation/2.0 http://schemas.opengis.net/citygml/transportation/2.0/transportation.xsd http://www.opengis.net/citygml/building/2.0 http://schemas.opengis.net/citygml/building/2.0/building.xsd http://www.opengis.net/citygml/cityobjectgroup/2.0 http://schemas.opengis.net/citygml/cityobjectgroup/2.0/cityObjectGroup.xsd http://www.opengis.net/citygml/tunnel/2.0 http://schemas.opengis.net/citygml/tunnel/2.0/tunnel.xsd http://www.opengis.net/citygml/cityfurniture/2.0 http://schemas.opengis.net/citygml/cityfurniture/2.0/cityFurniture.xsd http://www.opengis.net/citygml/bridge/2.0 http://schemas.opengis.net/citygml/bridge/2.0/bridge.xsd http://www.opengis.net/citygml/generics/2.0 http://schemas.opengis.net/citygml/generics/2.0/generics.xsd http://www.opengis.net/citygml/landuse/2.0 http://schemas.opengis.net/citygml/landuse/2.0/landUse.xsd">\n'
    lines
    with open(xml_data, "w") as f:
        f.writelines(lines)
    
    tree = ET.parse(xml_data)
    root = tree.getroot()
    
    def extract_element(element, xpath=''):
        
        element_list = []
        
        if element == 'bldg_id':
            for x in root.findall('{citygml}cityObjectMember'):
                bid = x.find('{bldg}Building')
                element_list.append(bid.attrib)

            df = pd.DataFrame()
            element_list = pd.Series(element_list)
            df[element] = element_list.values
            df[element] = df[element].astype(str)
        else:
            for x in root.findall('{citygml}cityObjectMember'):
                bid = x.find('{bldg}Building')
                for y in bid.findall(xpath):
                    try:
                        element_list.append(bid.attrib)
                        element_list.append(y.text)
                    except:
                        element_list.append(np.nan)
            
            bldg_id = pd.Series(element_list[::2])
            element_list = pd.Series(element_list[1::2])
            df = pd.DataFrame()
            df['bldg_id'] = bldg_id
            df['bldg_id'] = df['bldg_id'].astype(str)
            df[element] = element_list.values
            
        return df
    
    df_bldg_id = extract_element('bldg_id')
    df_measured_height = extract_element('height', './{bldg}measuredHeight')
    df_ground_polygon = extract_element('ground_polygon', './{bldg}boundedBy/{bldg}GroundSurface/{bldg}lod2MultiSurface/{gml}MultiSurface/{gml}surfaceMember/{gml}Polygon/{gml}exterior/{gml}LinearRing/{gml}posList')
    df_roof_polygon = extract_element('roof_polygon', './{bldg}boundedBy/{bldg}RoofSurface/{bldg}lod2MultiSurface/{gml}MultiSurface/{gml}surfaceMember/{gml}Polygon/{gml}exterior/{gml}LinearRing/{gml}posList')
    
    df = df_bldg_id.merge(df_measured_height, how='outer', on='bldg_id')
    df = df.merge(df_ground_polygon, how='outer', on='bldg_id')
    df = df.merge(df_roof_polygon, how='outer', on='bldg_id')
    
    return df
```

### Calculate roof height, area, generate ground polygons and set coordinate reference system
In this section the polygon values from the citygml data frame are transformed from 3D to 2D values. The height of the roof is calculated as the mean of all roofs of a building, the amount of roofs per building is calulated as well as the total area of all roofs per building. The z value is the elevation for each point in the polygon. The maximum z value of each roof polygon point is used as the roof height. The minimum of all elevation values per ground polygon is set as a new elevation column in the data. A geodataframe is created using geopandas and can also be plotted for inspection. The CRS is EPSG:25833 in the citygml files and can later be transformed into EPSG:4326 if needed.

```python
df_gml_1 = pd.read_csv('df_gml_polygons_unfininished.csv', sep=';')
```

```python
def get_roof_data(roof_polygon_column):
    try:
        poly_list = roof_polygon_column.split()
        elevation_list = poly_list[2::3]

        elevation_list = np.array(elevation_list)
        elevation_list = elevation_list.astype(np.float)
        del poly_list[2::3]
        elevation = np.max(elevation_list) # maximum of polygon z values of roof polygon points
        size=2
        poly_list = [poly_list[i:i+size] for i  in range(0, len(poly_list), size)]

        poly_list = list(tuple(map(float, x)) for x in poly_list)
        poly_list = Polygon(poly_list)
        poly_list = poly_list.area
        return pd.Series([poly_list, elevation])
    except:
        pass

df_gml_1[['roof_polygon', 'roof_elevation']] = df_gml_1.progress_apply(lambda x: get_roof_data(x['roof_polygon']), axis=1)

df_gml_1_group_roof_area = df_gml_1[['bldg_id', 'roof_polygon']].groupby(df_gml_1['bldg_id']).sum()
df_gml_1_group_roof_area = df_gml_1_group_roof_area.add_suffix('_area_sum').reset_index()

df_gml_1_groupd_roof_count = df_gml_1[['bldg_id', 'roof_polygon']].groupby(df_gml_1['bldg_id']).size()
df_gml_1_groupd_roof_count = df_gml_1_groupd_roof_count.reset_index(name='roof_count')


df_gml_1_group_roof_elevation = df_gml_1[['bldg_id', 'roof_elevation']].groupby(df_gml_1['bldg_id']).max()
df_gml_1_group_roof_elevation = df_gml_1_group_roof_elevation.add_suffix('_mean').reset_index()

df_group = df_gml_1_group_roof_area.merge(df_gml_1_group_roof_elevation, how='left', on='bldg_id')
df_group = df_group.merge(df_gml_1_groupd_roof_count, how='left', on='bldg_id')

df_gml_1.drop(['roof_polygon', 'roof_elevation'], axis=1, inplace=True)
df_gml_1 = df_group.merge(df_gml_1, how='left', on='bldg_id')
df_gml_1 = df_gml_1.drop_duplicates()
# df_gml_1['roof_elevation_mean'].mean()
```

```python
def get_polygon(polygon_column):
    try:
        poly_list = polygon_column.split()
        elevation_list = poly_list[2::3]

        elevation_list = np.array(elevation_list)
        elevation_list = elevation_list.astype(np.float)
        del poly_list[2::3]
        elevation = np.min(elevation_list) # minimum of polygon z values of ground polygon points
        size=2
        poly_list = [poly_list[i:i+size] for i  in range(0, len(poly_list), size)]

        poly_list = list(tuple(map(float, x)) for x in poly_list)
        poly_list = Polygon(poly_list)
        return pd.Series([poly_list, elevation])
    except:
        pass

df_gml_1[['ground_polygon', 'bldg_elevation']] = df_gml_1.progress_apply(lambda x: get_polygon(x['ground_polygon']), axis=1)

df_gml_1['calculated_height'] = df_gml_1['roof_elevation_mean'] - df_gml_1['bldg_elevation']

df_gml_1 = df_gml_1.rename({'ground_polygon':'geometry'}, axis='columns')
df_gml_1['geometry'] = geopandas.GeoSeries((df_gml_1['geometry']))
crs = {'init': 'epsg:25833'}
df_gml_1 = geopandas.GeoDataFrame(df_gml_1, crs=crs, geometry=df_gml_1['geometry'])
df_gml_1 = df_gml_1[df_gml_1['geometry'].notnull()]
df_gml_1.shape
```

```python
def dict_for_poi_grouping(poi_dataframe):
    columns = list(poi_dataframe)
    # remove lat_poi lon_poi and geometry
    columns = columns[2:-1]
    dictionary = {}
    for name in columns:
        dictionary.update({name: np.sum})

    return(dictionary)

dict_poi = dict_for_poi_grouping(pois)

df_pois_with_url = df_html_gml_buffer_join.groupby(['url']).agg(dict_poi)
df_pois_with_url = df_pois_with_url.reset_index()
```

Missing values are imputed using the median of each column with sklearn and the Imputer method. First the Imputer is specified, then fitted to the df_train dataset and can then be applied to train, test, validate sets with fit().

```python
impute_nans = Imputer(missing_values=np.nan, strategy='mean', axis=0)
imputed_df_train = impute_nans.fit(df_train)

def imputer_transform(df):
    imputed_df = pd.DataFrame(impute_nans.transform(df))
    imputed_df.columns = df.columns
    imputed_df.index = df.index
    df = imputed_df
    return df

df = imputer_transform(df)
df_train = imputer_transform(df_train)
#df_validate = imputer_transform(df_validate)
df_test = imputer_transform(df_test)
```

## Base models with all variables
Run base model with RandomForestRegressor to get variable importance, then run all models with needed features.

```python
def run_base_model(df_train, df_train_rent_sqm_log, df_test, df_test_rent_sqm_log, model, random=False):
    if not random:
        base_model = model()
    else:
        base_model = model(random_state = 1991)
    base_model.fit(df_train, df_train_rent_sqm_log)
    base_model_predictions = base_model.predict(df_test)
    base_model_predictions = pd.DataFrame(base_model_predictions)
    rmse = sqrt(mean_squared_error(df_test_rent_sqm_log, base_model_predictions))
    return base_model, base_model_predictions, rmse
```

```python
rf, rf_pred, rf_rmse = run_base_model(df_train, 
                                      df_train_rent_sqm_log, 
                                      df_test, 
                                      df_test_rent_sqm_log,
                                      RandomForestRegressor, 1991)
```

## Visualize Correlation between independent variables

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().unstack()#.abs()
    labels_to_drop = get_redundant_pairs(df)
    au_corr_pos = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    au_corr_neg = au_corr.drop(labels=labels_to_drop).sort_values(ascending=True)
    return au_corr_pos[0:n], au_corr_neg[0:n]

print("Top Absolute Correlations")
df_corr_pos, df_corr_neg = get_top_abs_correlations(df_train, 10)
df_corr_pos = pd.DataFrame(df_corr_pos)
df_corr_neg = pd.DataFrame(df_corr_neg)
df_corr = df_corr_pos.append(df_corr_neg)
df_corr = df_corr.round(3)
df_corr = df_corr.sort_values(by=[0], ascending=False)
df_corr
df_corr = df_corr.reset_index()
corr_list1 = df_corr['level_0'].tolist()
corr_list2 = df_corr['level_1'].tolist()
features_corr = list(set(corr_list1 + corr_list2))
df_corr.to_csv('ind_correlations.csv', index=False)
```

# Run the models
In this section, the models are run. The features can be specified in a dictionary, the models in a list called models. If the model has no random_state, please state None as the second item in the list in models. All models are run for each feature and then the rmse are added to a data frame. If new models are run, the column of the data frame has to be extended.

```python
feature_dict = {#'Rooms': features_room,
                'Real\nEstate\nAttributes': features_immo,
                '2D': features_2D,
                '3D': features_3D,
                'Spatial': features_spatial,
                '3D_spatial': features_3D_spatial,
                'POIs': features_poi,
               # 'variable\nimportance':features_variable_importance,
               # 'VIF':features_vif_all,
                'VIF&\nVariable\nImportance':features_vif_variable_importance,
               # 'all\nfeatures': features_all
}

models = [(LinearRegression, None),
          (RandomForestRegressor, 1991), 
          (SVR, None), 
          (GradientBoostingRegressor, 1991),
          (KNeighborsRegressor, None),
          (Ridge, 1991),
          (Lasso, 1991),
          (ElasticNet, 1991)]

rmse_feature_list = []

for key, value in tqdm_notebook(feature_dict.items()):
    for model, random in models:
        rmse = run_specific_models(value, model)
        rmse_feature_list.append(rmse)
        
        
df_rmse = pd.DataFrame(columns=['LinearRegression', 
                                'RandomForest', 
                                'SVR', 
                                'GradientBoosting', 
                                'KNeighbors',
                                'Ridge', 
                                'Lasso', 
                                'ElasticNet'])

feature_list = []
for key, value in feature_dict.items():
    feature_list.append(key)

df_rmse['index'] = pd.Series(feature_list).values
df_rmse = df_rmse.set_index('index')

df_rmse['LinearRegression'] = rmse_feature_list[0::len(models)] # linear regression
df_rmse['RandomForest'] = rmse_feature_list[1::len(models)] # random forest
df_rmse['SVR'] = rmse_feature_list[2::len(models)] # svr
df_rmse['GradientBoosting'] = rmse_feature_list[3::len(models)]
df_rmse['KNeighbors'] = rmse_feature_list[4::len(models)]# gra
df_rmse['Ridge'] = rmse_feature_list[5::len(models)]
df_rmse['Lasso'] = rmse_feature_list[6::len(models)]
df_rmse['ElasticNet'] = rmse_feature_list[7::len(models)]


df_test_rent_sqm_log_mean = df_test_rent_sqm_log
df_test_rent_sqm_log_mean = pd.DataFrame(df_test_rent_sqm_log_mean)
df_test_rent_sqm_log_mean['mean_rent_sqm_log'] = df_train_rent_sqm_log.mean()
mean_baseline = sqrt(mean_squared_error(df_test_rent_sqm_log, df_test_rent_sqm_log_mean['mean_rent_sqm_log']))
df_rmse.loc['Intercept'] = mean_baseline
# move mean_baseline to the top of the data frame
df_rmse = pd.concat([df_rmse.iloc[[(df_rmse.shape[0]-1)],:], df_rmse.drop('Intercept', axis=0)])
df_rmse
```
